## Reading Wikipedia table


```python
import pandas as pd
import numpy as np

link = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

tables = pd.read_html(link)
tor_db = tables[0].iloc[0:]
tor_db.columns = ['PostalCode','Borough','Neighbourhood']
tor_db
```

### Using geocoder

```python
import pandas as pd
import numpy as np

link = "https://cocl.us/Geospatial_data"
coord = pd.read_csv(link)
coord.columns = ['PostalCode', 'Latitude','Longitude']
coord = pd.DataFrame(coord)
tor_db = pd.DataFrame(tor_db)

coord_db = pd.merge(tor_db, coord, on='PostalCode', how='inner')
coord_db
```
